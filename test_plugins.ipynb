```python
import folium
from folium import plugins

# folium.Map(location=[36.109482, 140.103939],control_scale=True)
```

```python
# Dual Map
dualmap = plugins.DualMap(location=[36.109482, 140.103939])
folium.Marker([36.109482, 140.103939]).add_to(dualmap.m1)
dualmap
```

```python
#  Locate control
map_locatecontrol = folium.Map(location=[36.109482, 140.103939],control_scale=True)
plugins.LocateControl().add_to(map_locatecontrol )
map_locatecontrol
```

```python
map_tsgeojson = folium.Map(location=[36.109482, 140.103939],control_scale=True)
plugins.TimestampedGeoJson({
    'type' : 'FeatureCollection',
    'features' : [{
        'type' : 'Feature',
        'geometry' : {
            'type' : 'LineString',
            'coordinates' : [[-70,-25],[-70,35],[70,35]],
        },
        'properties' : {
            'times': [1435708800000, 1435795200000, 1435881600000]
        }
    }]
}).add_to(map_tsgeojson)

map_tsgeojson
```
